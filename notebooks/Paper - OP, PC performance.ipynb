```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import seaborn as sns

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import extra_scripts.plot_functions as pf
```

```python
LINEWIDTH = 5
FIG_SIZE = (8, 7)


def plot_bar(value_lists, xlabel=None, fig_size=FIG_SIZE,
        fig_pos=None, ylabel=None, xticklabel=None,
        xlabel_rotation=None, width=-1, gap=1, yticklabel=None, legend=None,
        errorbar_list=None, color_list=sns.color_palette(n_colors=10), ylim=None, ncol=1,
        handlelength=None, loc=0, bbox_to_anchor=None,
        handletextpad=None, columnspacing=None, vlines=None, hlines=None,
        hatches=None):
    fig, ax = pf.start_plotting(fig_size, fig_pos)
    N = len(value_lists[0])
    if width < 0:
        width = 0.75 * gap / len(value_lists)
    ind = np.arange(N) * gap
    rects = []
    for i in range(len(value_lists)):
        rect = ax.bar(ind, value_lists[i], width, color=color_list[i],
                yerr=errorbar_list[i] if errorbar_list else None, error_kw={"ecolor": "black"},
                hatch=None if not hatches else hatches[i])
        ind = ind + width
        rects.append(rect)
    xlim = (-width, max(ind))
    
    x = np.array([r.get_x() for r in ax.patches])
    w = np.array([r.get_width() for r in ax.patches])

    ax.hlines([op_rand, pc_rand, both_rand], x, x + w, color="grey", linestyle="--")

    if legend:
        legend_xgb = ax.legend(rects, legend, loc=loc,
                    bbox_to_anchor=bbox_to_anchor,
                    ncol=ncol, handlelength=handlelength,
                    handletextpad=handletextpad,
                    columnspacing=columnspacing,
                    frameon=False)
        legend_rand = [Line2D([0], [0], marker=' ', linestyle="--", color='grey', label='Random')]
        ax.legend(handles=legend_rand, loc=(0.475, 0.72),
                frameon=False, ncol=1, handletextpad=0.5)
        ax.add_artist(legend_xgb)
    
    
    pf.end_plotting(fig, ax, xlabel=xlabel, ylabel=ylabel,
            xticklabel=(ind - width * len(value_lists), xticklabel) if xticklabel else None,
            xlabel_rotation=xlabel_rotation, xlim=xlim,
            yticklabel=yticklabel, ylim=ylim)
    return fig
```

```python
# [all, stopwords, content words]

op_rand = [0.11096, 0.173996, 0.08092]
pc_rand = [0.1429720, 0.18570269, 0.1270957]
both_rand = [0.21493012, 0.22953, 0.19117901]

op = [0.34665, 0.466914, 0.1582359]
pc = [0.2710990, 0.424338, 0.1621139]
both = [0.630302, 0.685531, 0.47348707]
# orange dbc257
# blue # 4264ed
palette = sns.color_palette(n_colors=10)
color_list = palette
hatches = ['//', '\\', 'x', '+']
fig = plot_bar([op, pc, both],
                  ylabel="F1 Score", xticklabel=["All", "Stopwords", "Content words"],
                  xlabel_rotation=20, legend=["OP only", "PC only", "OP \\& PC"],
                  hlines=[0.3, 0.5, 0.5],
                  loc="upper right", ncol=2, bbox_to_anchor=(1.045,1.05), columnspacing=0.4,
                  hatches=hatches, color_list=color_list, ylim=(0, 0.95), handletextpad=0.4)
fig.savefig("../records/plots/op-pc-perf.png", bbox_inches="tight")
```
